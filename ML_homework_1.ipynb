```python
# ДЗ №1 по ML 
# Предсказать параметр MEDV из датасета housing.csv:
# 1 - модель линейной регрессии
# 2 - разбить train/test на 0.8/0.2
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
random_seed = 123
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum().sum()
```

```python
# Значения по оси X
df.drop('MEDV', axis=1)
```

```python
# Значения по оси Y
df['MEDV']
```

```python
# разбиваем на тренировочный сет и на тестовый сет в отношении 0.8/0.2 соответственно
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis=1),
                                                   df['MEDV'],
                                                   test_size=0.2,
                                                   random_state = random_seed) 
```

```python
df.shape
```

```python
X_train
```

```python
# модель линейной регрессии
model = LinearRegression()
```

```python
# обучаем модель
model.fit(X_train, y_train)
```

```python
# предсказываем значения для тестового сета
test_prediction = model.predict(X_test)
```

```python
test_prediction
```

```python
# предсказываем значения MEDV для всего df
medv_prediction = model.predict(df.drop('MEDV', axis=1))
```

```python
df['MEDV_prediction'] = medv_prediction
```
